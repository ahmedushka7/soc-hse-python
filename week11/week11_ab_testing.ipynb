## Введение в А/Б-тестирование

> "Making decisions is easy — what’s hard is making the right decisions."

### 1. Идея 

Каждый день мы используем кучу приложений (Yandex Taxi, Google Maps, Avito, YouTube). Какие-то из них кажутся нам довольно удобными и привлекательными, а какие-то нет. Обычно мы не задумываемся, что именно делает приложение привлекательным, а иногда, даже начав задумываться об этом, мы не можем определить четкие критерии, мы лишь опираемся на наши ощущения.


Теперь представьте, что вы выступаете создателем нового приложения (нового продукта) и конечно же вы заинтересованы в том, чтобы сделать его удобным для пользователей. Но как вы определите это? Мнение каждого субъективно и то, что кажется привлекательным для одних, совсем не будет казаться таковым для других.


**Разумный ответ: Спроси своих пользователей**


Конечно же мы не будем обзванивать каждого пользователя и спрашивать у него про новый дизайн, это слишком дорого для нас. Но мы можем использовать статистику и **провести А/Б-тест**.

### 2. Определение

__Все начинается с гипотезы__

Предположим, что мы хотим проверить, какой дизайн логотипа Google более привлекателен для пользователя – черно-белый или цветной.

Чтобы это проверить, мы должны:

1. Взять выборку наших пользователей (обычно случайно и равновероятно)
2. Разделить их на 2 группы: _контроль_ и _тест_
3. На стороне разработки приложения сделать так, чтобы пользователи из группы А (контроль) видели цветной дизайн, а из группы Б (тест) – черно-белый
4. На протяжении какого-то времени собирать статистику для каждого пользователя по определенной метрике
5. Сагрегировать метрику для всех пользователей из 2х групп и получить 2 статистики (например, средние)
6. Проверить статистическим тестом, в какой группе метрика больше
